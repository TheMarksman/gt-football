# Getting Started

```python
import numpy as np
import pandas as pd
```

```python
data20 = pd.read_csv('cfb20.csv', encoding='latin-1')
data19 = pd.read_csv('cfb19.csv', encoding='latin-1')
data18 = pd.read_csv('cfb18.csv', encoding='latin-1')
data17 = pd.read_csv('cfb17.csv', encoding='latin-1')
data16 = pd.read_csv('cfb16.csv', encoding='latin-1')
data15 = pd.read_csv('cfb15.csv', encoding='latin-1')
data14 = pd.read_csv('cfb14.csv', encoding='latin-1')
data13 = pd.read_csv('cfb13.csv', encoding='latin-1')
```

```python
data19
```

```python
TEAM = 'Georgia Tech (ACC)'

def filter_by_team(team_string, df):
    is_team = df['Team'] == team_string
    return df[is_team]

team_data20 = filter_by_team(TEAM, data20)
team_data19 = filter_by_team(TEAM, data19)
team_data18 = filter_by_team(TEAM, data18)
team_data17 = filter_by_team(TEAM, data17)
team_data16 = filter_by_team(TEAM, data16)
team_data15 = filter_by_team(TEAM, data15)
team_data14 = filter_by_team(TEAM, data14)
team_data13 = filter_by_team(TEAM, data13)
```

```python
team_data20.insert(1, 'Year', 2020, False)
team_data19.insert(1, 'Year', 2019, False)
team_data18.insert(1, 'Year', 2018, False)
team_data17.insert(1, 'Year', 2017, False)
team_data16.insert(1, 'Year', 2016, False)
team_data15.insert(1, 'Year', 2015, False)
team_data14.insert(1, 'Year', 2014, False)
team_data13.insert(1, 'Year', 2013, False)
```

```python
frames = [team_data20, team_data19, team_data18, team_data17, team_data16, team_data15, team_data14, team_data13]
```

```python
team_combined_data = pd.concat(frames)
team_combined_data.drop(team_combined_data.columns[0], axis=1)
```

```python
def camel(snake_str):
    first, *others = snake_str.split('_')
    return ''.join([first.lower(), *map(str.title, others)])

columnMap = {}

for column in team_combined_data.columns[1:]:
    parsedColumn = column.replace('.', '_')
    parsedColumn = camel(parsedColumn)
    columnMap[column] = parsedColumn
    
print(columnMap)

team_combined_data = team_combined_data.rename(columns=columnMap)
    
```

```python
team_combined_data.to_json('gt_football_data.json', orient='records')
```
